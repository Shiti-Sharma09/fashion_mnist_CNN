```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import os

tf.random.set_seed(42)
np.random.seed(42)

print("TensorFlow version:", tf.__version__)
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")
print(f"Class names: {class_names}")
```

```python
def plot_sample_images(x_data, y_data, class_names, num_samples=12):
    fig, axes = plt.subplots(3, 4, figsize=(12, 9))
    fig.suptitle('Sample Images from Fashion-MNIST Dataset', fontsize=16)

    for i in range(num_samples):
        row, col = i // 4, i % 4
        axes[row, col].imshow(x_data[i], cmap='gray')
        axes[row, col].set_title(f'{class_names[y_data[i]]}')
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()

plot_sample_images(x_train, y_train, class_names)
```

```python
def plot_class_samples(x_data, y_data, class_names):

    fig, axes = plt.subplots(2, 5, figsize=(15, 8))
    fig.suptitle('Sample Images from Each Class - Fashion-MNIST Dataset', fontsize=16)
    
    # Find first occurrence of each class
    class_indices = {}
    for i, label in enumerate(y_data):
        if label not in class_indices:
            class_indices[label] = i
        if len(class_indices) == 10:  # All classes found
            break
    
    # Plot one image from each class
    for class_idx in range(10):
        row, col = class_idx // 5, class_idx % 5
        image_idx = class_indices[class_idx]
        
        axes[row, col].imshow(x_data[image_idx], cmap='gray')
        axes[row, col].set_title(f'Class {class_idx}: {class_names[class_idx]}', 
                               fontsize=12, fontweight='bold')
        axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.show()

# Plot random sample images
plot_class_samples(x_train, y_train, class_names)
```

```python
def plot_class_distribution(y_data, class_names, title):
    unique, counts = np.unique(y_data, return_counts=True)
    plt.figure(figsize=(12, 6))
    bars = plt.bar(class_names, counts, color='skyblue', edgecolor='navy', alpha=0.7)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel('Classes')
    plt.ylabel('Samples per Class')
    plt.xticks(rotation=45, ha='right')
    for bar, count in zip(bars, counts):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100, str(count), ha='center')
    plt.tight_layout()
    plt.show()

plot_class_distribution(y_train, class_names, 'Training Data - Class Distribution')
```

```python
x_train_norm = x_train.astype('float32') / 255.0
x_test_norm = x_test.astype('float32') / 255.0

x_train_reshaped = x_train_norm.reshape(-1, 28, 28, 1)
x_test_reshaped = x_test_norm.reshape(-1, 28, 28, 1)

y_train_sparse = y_train
y_test_sparse = y_test

print(f"x_train: {x_train_reshaped.shape}, y_train: {y_train_sparse.shape}")
```

```python
def create_cnn_model(input_shape=(28, 28, 1), num_classes=10):
    """Create a CNN model for Fashion-MNIST classification"""
    model = keras.Sequential([
        # Input layer (fixes the warning)
        layers.Input(shape=input_shape),
        
        # First Convolutional Block
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        
        # Second Convolutional Block
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        
        # Third Convolutional Block
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.25),
        
        # Flatten and Dense layers
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

model = create_cnn_model()
model.summary()
```

```python
model.compile(
    optimizer='adamw',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
datagen.fit(x_train_reshaped)
```

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7)
]
```

```python
EPOCHS = 30
BATCH_SIZE = 128

history = model.fit(
    datagen.flow(x_train_reshaped, y_train_sparse, batch_size=BATCH_SIZE),
    epochs=EPOCHS,
    validation_data=(x_test_reshaped, y_test_sparse),
    callbacks=callbacks,
    verbose=1
)
```

```python
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test_sparse, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
predictions = model.predict(x_test_reshaped)
predicted_classes = np.argmax(predictions, axis=1)
```

```python
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.plot(history.history['accuracy'], label='Training Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Accuracy')
    ax1.legend()

    ax2.plot(history.history['loss'], label='Training Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Loss')
    ax2.legend()

    plt.show()

plot_training_history(history)
```

```python
### Confusion Matrix

def plot_confusion_matrix(y_true, y_pred, class_names):
    """Plot confusion matrix"""
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

plot_confusion_matrix(y_test, predicted_classes, class_names)
```

```python
print(classification_report(y_test, predicted_classes, target_names=class_names))
```

```python
### Prediction Visualization


def plot_predictions(x_test, y_test, predictions, class_names, num_samples=12):
    """Plot predictions with actual vs predicted labels"""
    fig, axes = plt.subplots(3, 4, figsize=(15, 12))
    fig.suptitle('Predictions vs Actual Labels', fontsize=16)
    
    for i in range(num_samples):
        row, col = i // 4, i % 4
        
        # Get prediction
        pred_class = np.argmax(predictions[i])
        actual_class = y_test[i]
        confidence = predictions[i][pred_class]
        
        # Plot image
        axes[row, col].imshow(x_test[i].reshape(28, 28), cmap='gray')
        
        # Set title with prediction and confidence
        if pred_class == actual_class:
            color = 'green'
            title = f' {class_names[pred_class]}\n({confidence:.2f})'
        else:
            color = 'red'
            title = f' Pred: {class_names[pred_class]}\nActual: {class_names[actual_class]}\n({confidence:.2f})'
        
        axes[row, col].set_title(title, color=color, fontsize=10, fontweight='bold')
        axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.show()

plot_predictions(x_test_reshaped, y_test, predictions, class_names)
```

```python
def predict_single_image(model, image, class_names):
    image_norm = image.astype('float32') / 255.0
    image_reshaped = image_norm.reshape(1, 28, 28, 1)
    prediction = model.predict(image_reshaped, verbose=0)
    predicted_class = np.argmax(prediction)
    confidence = prediction[0][predicted_class]
    return class_names[predicted_class], confidence

# Test multiple random images
num_random_tests = 5
for i in range(num_random_tests):
    random_index = np.random.randint(0, len(x_test))
    test_image = x_test[random_index]
    
    predicted_class, confidence = predict_single_image(model, test_image, class_names)
    actual_class = class_names[y_test[random_index]]
    
    print(f"Test {i+1} - Index: {random_index}")
    print(f"Predicted: {predicted_class} ({confidence:.4f})")
    print(f"Actual: {actual_class}")
    print(f"Correct: {'' if predicted_class == actual_class else ''}")
    print("-" * 40)
```
